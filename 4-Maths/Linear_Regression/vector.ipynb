```python
# Dot Product of 2 vectors 
a = np.array([2, 3])
b = np.array([4, 4])

a_dot_b = np.dot(a,b)
a_cross_b = np.cross(a,b)
```

```python
# projection 

a = np.array([2, 5])

v = np.array([8, -6])

# magnitude of v vector
mag_v = np.sqrt(np.sum(v**2))

proj_a_on_v = (np.dot(a,v)/mag_v**2)*v

print(proj_a_on_v)
```
