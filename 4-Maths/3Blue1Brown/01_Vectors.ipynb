```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1 = np.array([1, 2])
v2 = np.array([3, 4])
```

```python
r_add = v1 + v2 
```

```python
v1 - v2
```

```python
v1 * v2
```

```python
v1 + v2 
```

```python
plt.quiver(0,0 , v1[0], v1[1],angles='xy', scale_units='xy', scale=1, color='r')
plt.quiver(0,0 , v2[0], v2[1],angles='xy', scale_units='xy', scale=1, color='b')
plt.quiver(0,0 , r_add[0], r_add[1],angles='xy', scale_units='xy', scale=1, color='g')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.grid()
```

```python
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='b', alpha=0.3)
plt.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', alpha=0.3)

plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid()
plt.legend()
plt.show()
```
