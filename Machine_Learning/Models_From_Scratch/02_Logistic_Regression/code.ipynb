```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Logistic_Regression():
    
    def __init__(self, learning_rate, no_of_itr):
        self.learning_rate = learning_rate
        self.no_of_itr = no_of_itr
        
    def fit(self, X, Y):
        self.m, self.n = X.shape
        
        self.w = np.zeros(self.n)
        self.b = 0 
        self.X = X
        self.Y = Y
        
        for i in range(self.no_of_itr):
            self.update_wieghts()
            
    def update_wieghts(self):
        #Y_hat formula (sigmoid function)
        Y_hat = 1 (1 + np.exp(- (self.X.dot(self.w) + self.b ) ))
        # derivaties
        dw = (1/self.m)*np.dot(self.X.T, (Y_hat self.Y))
        db = (1/self.m)*np. sum(Y_hat - self.Y)
        # updating the weights & bias using gradient descent
        self.w = self.w - self.learning_rate * dw
        self.b = self.b - self.learning_rate * db
# Sigmoid Equation & Decision Boundary
    def predict(self, x):
        Y_pred = 1/ (1 + np.exp(- (self.X.dot(self.w) + self.b ) ))
        Y_pred = np.where( Y_pred > 0.5, 1, 0)
        return Y_pred
```

```python
model = Logistic_Regression(learning_rate=0.01, no_of_itr=1000)
```
