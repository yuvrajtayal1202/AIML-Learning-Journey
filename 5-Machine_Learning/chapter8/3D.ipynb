```python
from sklearn.datasets import make_swiss_roll
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting
import matplotlib.pyplot as plt
# ðŸŒ€ Make 3D twisty data (like a cinnamon roll)
X_3D, t = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)

print("Shape of 3D data:", X_3D.shape)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_3D[:, 0], X_3D[:, 1], X_3D[:, 2], c=t, cmap=plt.cm.Spectral)
ax.set_title("ðŸ”· Original 3D Swiss Roll")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```

```python
from sklearn.decomposition import PCA


pca_3d = PCA(n_components=2)
X_2D_flat = pca_3d.fit_transform(X_3D)

print("Flattened shape:", X_2D_flat.shape)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_2D_flat[:, 0], X_2D_flat[:, 1], c=t, cmap=plt.cm.Spectral)
plt.title("ðŸ§» PCA - Swiss Roll Flattened to 2D")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.show()
```
