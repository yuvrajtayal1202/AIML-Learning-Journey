```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# ðŸŒ¸ The Iris dataset has 150 flowers with 4 features (columns)
# Features: sepal length, sepal width, petal length, petal width

iris = load_iris()
X = iris.data          # flower data
y = iris.target        # flower types (0, 1, 2)
labels = iris.target_names

print("Shape of data:", X.shape)  # (150, 4)
```

```python
# Let's reduce it to 2D using PCA so we can draw it!

pca = PCA(n_components=2)
X_2D = pca.fit_transform(X)

print("New shape:", X_2D.shape)  # (150, 2)
```

```python
# ðŸŽ¨ Scatter plot of flowers in 2D

plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_2D[:, 0], X_2D[:, 1], c=y, cmap="viridis")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("ðŸŒ¸ PCA - Iris Flower Data (2D)")
plt.legend(handles=scatter.legend_elements()[0], labels=labels.tolist())
plt.grid(True)
plt.show()
```

```python
# This tells us how much of the original data is preserved
# e.g., 0.92 means 92% of the information is kept in the 1st axis

print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total info kept:", np.sum(pca.explained_variance_ratio_) * 100, "%")
```

```python
# Let's try just 1 dimension
pca_1d = PCA(n_components=1)
X_1D = pca_1d.fit_transform(X)

print("1D shape:", X_1D.shape)

# Plotting in 1D (as a strip)
plt.scatter(X_1D, np.zeros_like(X_1D), c=y, cmap="viridis")
plt.title("PCA - 1D View")
plt.yticks([])
plt.xlabel("PCA 1")
plt.grid(True)
plt.show()
```
