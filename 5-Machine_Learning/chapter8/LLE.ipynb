```python
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt

# Make Swiss Roll data
X, color = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)
```

```python
# Apply Locally Linear Embedding (LLE)
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_lle = lle.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X_lle[:, 0], X_lle[:, 1], c=color, cmap=plt.cm.Spectral)
plt.title("🎯 LLE - Swiss Roll Unfolded with Local Friend Groups")
plt.grid(True)
plt.show()
```

```python
# Isomap
from sklearn.manifold import Isomap
from sklearn.datasets import make_swiss_roll
import matplotlib.pyplot as plt

# Make 3D Swiss Roll
X, color = make_swiss_roll(n_samples=1000, noise=0.1, random_state=42)

# Apply Isomap
iso = Isomap(n_components=2, n_neighbors=10)
X_iso = iso.fit_transform(X)

# Plot
plt.scatter(X_iso[:, 0], X_iso[:, 1], c=color, cmap=plt.cm.Spectral)
plt.title("🗺️ Isomap - Swiss Roll Unfolded")
plt.grid(True)
plt.show()
```
